```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data=pd.DataFrame({"Satan_ID" : range(1,6), "Satış_Hedefi" : [100000,125000,110000,145000,130000], "Gerçekleşen_Satış": [99000,130000,95000,150000,80000]})
data
```

```python
data["Gerçekleşme_Oranı"]=round(data["Gerçekleşen_Satış"] / data["Satış_Hedefi"],2)
data
```

```python
def komisyon(x):
  if x<=.90:
    return .02
  if x<=.99:
    return .03
  else:
    return .04

data["Komisyon_Oranı"]=data["Gerçekleşme_Oranı"].apply(komisyon) 
data
```

```python
data["Komisyon_Tutarı"]=data["Komisyon_Oranı"]*data["Gerçekleşen_Satış"]
data
```
